```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Dados/Mall_Customers.csv')
data.shape
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
# configurar a visualização
#%matplotlib inline
%config InlineBackend.figure_format = 'svg'
#configurando parâmetros das fontes
plt.rcParams['font.family'] = 'sans-serif'
```

```python
sns.pairplot(data);
```

```python
plt.figure(figsize=(11,6))
mask = np.triu(np.ones_like(data.corr(), dtype=np.bool))
sns.heatmap(data.corr(), annot = True, cmap = 'copper', mask=mask)
plt.title('Correlação entre os Dados', fontsize = 15);
```

# Descritivo

```python
data.describe()
```

```python
data.describe(include = 'object')
```

```python
plt.figure(figsize= (11, 6))

plt.subplot(1, 2, 1)
sns.set(style = 'whitegrid')
sns.distplot(data['Annual Income (k$)'])
plt.title('Distributição Anual', fontsize = 20)
plt.xlabel('Renda')
plt.ylabel('Quantidade')

plt.subplot(1, 2, 2)
sns.set(style = 'whitegrid')
sns.distplot(data['Age'], color = 'red')
plt.title('Distribuição de Idade', fontsize = 20)
plt.xlabel('Idade')
plt.ylabel('Quantidade')
```

```python
labels = ['Female', 'Male']
size = data['Gender'].value_counts()
colors = ['lightgreen', 'orange']
explode = [0, 0.001]

plt.figure(figsize = (9, 9))
plt.pie(size, colors = colors, explode = explode, labels = labels, shadow = True, startangle = 90, autopct = '%.2f%%')
plt.title('Genero', fontsize = 20)
plt.axis('off')
plt.legend();
```

```python
plt.figure(figsize = (13, 6))
sns.countplot(data['Age'], palette = 'hsv')
plt.title('Distribuição de Idade', fontsize = 20);
```

```python
plt.figure(figsize = (12, 6))
sns.distplot(data['Annual Income (k$)'], color = 'red')
plt.title('Distribuição de Renda', fontsize = 20);
```

```python
plt.figure(figsize = (12, 6))
sns.distplot(data['Spending Score (1-100)'], color = 'black')
plt.title('Distribuição de Pontação de Gastos', fontsize = 20);
```

```python
plt.figure(figsize = (12, 6))
sns.boxenplot(data['Gender'], data['Spending Score (1-100)'], palette = 'Blues')
plt.title('Genero vs Pontuação de Gastos', fontsize = 20);
```

```python
plt.figure(figsize = (12,6))
sns.violinplot(data['Gender'], data['Annual Income (k$)'], palette = 'rainbow')
plt.title('Genero vs Renda Anual', fontsize = 20);
```

```python
plt.figure(figsize= (12, 6))
sns.stripplot(data['Gender'], data['Age'], palette = 'Purples', size = 10)
plt.title('Genero vs Idade', fontsize = 20);
```

```python
x = data['Annual Income (k$)']
y = data['Age']
z = data['Spending Score (1-100)']

plt.figure(figsize=(12,6))
sns.lineplot(x, y, color = 'blue')
sns.lineplot(x, z, color = 'pink')
plt.title('Renda Anual vs Idade e Pontuação Gastos', fontsize = 20);
```
